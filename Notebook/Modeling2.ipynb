### **Multiple Linear Regression**

Linear regression, often known as simple regression, creates a relationship between two variables. Linear regression is represented visually as a straight line, with the slope determining how a change in one variable affects a change in the other. A linear regression relationship's y-intercept reflects one variable's value when the other's value is 0.

**Multiple linear regression** estimates the relationship between two or more independent variables and one dependent variable.

$$\hat y = \beta_0 + \beta_1x_1 + \dots + \beta_px_p + \epsilon$$

where $p$ is the number of features (independent variables) in the model.

Here’s what each term means:

* **$\hat{y}$**: This is the predicted value of the dependent variable (y). It’s what we’re trying to estimate or predict.
* **$\beta_0$**: This is the intercept, or the predicted value of $y$ when all independent variables ($x_1$, $x_2$, ..., $x_p$) are 0. It's like the starting point of our prediction.
* **$\beta_1$**: This is the regression coefficient for the first independent variable ($x_1$). It tells we how much $y$ will change when $x_1$ increases or decreases by 1 unit, assuming all other variables stay the same.
* **$\beta_p$**: This is the regression coefficient for the last independent variable ($x_p$). Just like $\beta_1$, it shows how much $y$ will change when $x_p$ increases or decreases by 1 unit, keeping everything else constant.
* **$x_1$, $x_2$, ..., $x_p$**: These are the independent variables (also called predictors or features). They help predict the value of $y$.
* **$\epsilon$**: This is the error term. It represents the difference between the actual value of $y$ and the predicted value $\hat{y}$. In simple terms, it’s what we missed or couldn’t explain with our model.

### **Example**:

Let’s say we are trying to predict the price of a house. The price ($y$) depends on several factors, like:

1. **Size of the house** ($x_1$) (in square feet)
2. **Number of bedrooms** ($x_2$)
3. **Age of the house** ($x_3$) (in years)

Using multiple linear regression, the equation could look like this:

$$ \hat y = 50,000 + 200x_1 + 10,000x_2 - 500x_3 + \epsilon $$

Here’s how to interpret it:

- **$50,000$** is the intercept. Even if the house is tiny, has no bedrooms, and is brand new, the model starts at $50,000$.
- **$200x_1$**: For every additional square foot of the house, the price increases by $200$.
- **$10,000x_2$**: For every additional bedroom, the price increases by $10,000$.
- **$-500x_3$**: For every year older the house is, the price decreases by $500$.
- **$\epsilon$**: This accounts for any variation in house prices not explained by the model, like neighborhood or market conditions.

In this way, multiple linear regression helps we predict outcomes (like house prices) based on several factors (like size, bedrooms, and age).

### **Errors in Regression**

When we fit a regression line to data, it tries to get as close as possible to the actual data points. However, it rarely predicts the exact values, so there will always be some errors. These errors represent the difference between the actual value and the predicted value ($y - \hat{y}$). Let's explore some common types of errors used to evaluate the performance of regression models:

---

### **Mean Absolute Error (MAE)**

The **Mean Absolute Error (MAE)** is one of the simplest ways to measure errors in regression. It calculates the difference between the actual values and predicted values, takes the absolute value (so we don’t have to worry about negative and positive errors canceling each other out), and then averages these differences. It tells us, on average, how much our model's predictions deviate from the actual values.

$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

- **Example**: If we are predicting house prices and the MAE is 10,000, this means that, on average, the predicted house price is off by $10,000 from the actual price.

---

### **Mean Squared Error (MSE)**

The **Mean Squared Error (MSE)** is similar to MAE, but instead of taking the absolute differences, it squares the differences before averaging them. This squaring makes MSE more sensitive to large errors (outliers). If there are big differences between the actual and predicted values, MSE will heavily penalize those errors.

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

- **Example**: If the predicted house price is way off for some data points (e.g., predicting $200,000$ for a house that’s actually worth $400,000$), MSE will give more weight to those large errors because of the squaring.

---

### **Root Mean Squared Error (RMSE)**

The **Root Mean Squared Error (RMSE)** is simply the square root of the MSE. We use RMSE because MSE gives us squared units, which can be difficult to interpret. RMSE puts the error back into the original units of the data, making it easier to understand.

$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

- **Example**: If we are predicting house prices in dollars, RMSE will also be in dollars, which makes it easier to interpret. If RMSE is 12,000, it means the typical error in prediction is around $12,000.

---

### **Mean Absolute Percentage Error (MAPE)**

The **Mean Absolute Percentage Error (MAPE)** is similar to MAE but expressed as a percentage. Instead of showing the error in terms of the units of the data (e.g., dollars), MAPE shows the error as a percentage of the actual values. It’s useful when we want to understand the average error in terms of percentage, making it easier to explain to non-technical audiences.

$$MAPE = \frac{100\%}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|$$

- **Example**: If we’re predicting house prices and the MAPE is 5%, this means that, on average, our predictions are off by 5% of the actual price.

---

### **Summary**

- **MAE**: Simple average of absolute errors. Good for understanding the overall average error.
- **MSE**: Squares the errors, making it more sensitive to large errors (outliers).
- **RMSE**: The square root of MSE, making it easier to interpret in the original units.
- **MAPE**: Expresses the error as a percentage, making it easy to explain in business terms.

Each error metric provides different insights, and choosing the right one depends on our specific needs. If we care more about large errors, MSE or RMSE is useful. If we want an intuitive, straightforward measure, MAE or MAPE works well.

### When to Use Different Regression Error Metrics

Choosing the right error metric depends on the context of our problem and what we care about in terms of accuracy. 

---

### **1. Mean Absolute Error (MAE)**

**Best for**: When we want a simple and easy-to-interpret measure of average error, and we care equally about all errors (small and large).

- **Scenario**: Predicting rental prices for apartments in a city. If we want to know, on average, how far off our predictions are from actual prices without worrying too much about large errors (like predicting a luxury apartment too low), MAE is a good metric.
  - **Why**: It gives we a straightforward average of the errors, making it easy to explain to stakeholders (e.g., “on average, we’re off by $200 per month”).

---

### **2. Mean Squared Error (MSE)**

**Best for**: When we care more about large errors and want to heavily penalize large deviations from the actual value.

- **Scenario**: Predicting the costs of medical treatments. Suppose some treatments are very expensive (e.g., surgeries) and we want to avoid large prediction errors for these high-cost treatments.
  - **Why**: MSE squares the errors, so large prediction errors (like predicting $5,000 when the actual cost is $10,000) will have a much bigger impact on the error score. This makes it better for use cases where large errors are particularly costly or dangerous.

---

### **3. Root Mean Squared Error (RMSE)**

**Best for**: When we want to penalize large errors (like MSE) but also need the error metric in the same units as the predicted value, making it easier to interpret.

- **Scenario**: Predicting house prices in dollars. Suppose we are predicting house prices and want a metric that penalizes large errors but is also in terms of dollars (so we can explain the error size clearly).
  - **Why**: RMSE gives we a measure in the same units as our data (e.g., dollars), so it’s easier to interpret. If RMSE is $10,000, we can say, “Our average prediction is off by about $10,000.”

---

### **4. Mean Absolute Percentage Error (MAPE)**

**Best for**: When we need to explain errors in percentage terms, or when the scale of the dependent variable varies significantly.

- **Scenario**: Forecasting monthly sales for different stores. If Store A typically makes $1,000 in sales and Store B makes $100,000, the absolute error for Store A and Store B would be very different. However, if we care more about the relative size of the error (e.g., how far off we are as a percentage of the sales), MAPE is a good choice.
  - **Why**: MAPE expresses the error as a percentage of the actual values, so it’s easier to compare across different scales. If MAPE is 10%, it means our prediction is, on average, off by 10% of the actual sales.

---

### **Choosing the Right Metric**

- **Use MAE** when we want a simple and intuitive measure of average error that doesn’t heavily penalize large errors.
- **Use MSE** when large errors are much more important or costly, and we want to emphasize minimizing those.
- **Use RMSE** when we want to penalize large errors, but still need to interpret the error in the same units as our data.
- **Use MAPE** when we need to explain the error as a percentage, making it easier to understand relative to the size of the actual values. It’s useful for comparison across different scales.

Each metric has its strengths, and our choice should depend on the problem at hand and what we’re trying to prioritize—whether it's simplicity, dealing with outliers, or explaining the error in understandable terms.

### **Model Comparisons**

When comparing regression models, some key metrics help us understand how well the model fits the data. Let's break down two of the most important ones: **$R^2$** and **Adjusted $R^2$**.

---

### **1. $R^2$ (Coefficient of Determination)**

**What it does**: $R^2$ tells us how much of the variation in the dependent variable (the thing we're trying to predict) is explained by the independent variables (the predictors). It's a number between 0 and 1:

- **0**: Means the model explains none of the variance (i.e., it doesn’t predict anything useful).
- **1**: Means the model perfectly explains all the variance in the dependent variable.

**Formula**:
$$ R^2 = 1 - \frac{SSE}{TSS} $$

- **SSE** (Sum of Squared Errors) is the error between the predicted and actual values.
- **TSS** (Total Sum of Squares) is the total variation in the actual data.

---

**Example**:
- **Scenario**: We are predicting the sales of ice cream based on temperature.
- If $R^2 = 0.75$, it means 75% of the variation in ice cream sales is explained by the temperature, and the remaining 25% is due to other unknown factors.

---

### **2. Adjusted $R^2$**

**Why we need it**: While $R^2$ gives us a general idea of how well a model fits, it has a flaw—$R^2$ always increases when we add more independent variables (even if they don't really help). This can make it seem like a model is better just because it has more variables, which isn't always true.

**Adjusted $R^2$** fixes this problem by penalizing the model for adding more predictors that don’t improve the fit enough.

**Formula**:
$$ \text{Adjusted }  R^2 = 1 - \frac{(1 - R^2)(N - 1)}{N - p - 1} $$

- **N**: Number of data points (samples).
- **p**: Number of predictors (independent variables).

---

**Example**:
- **Scenario**: We add several extra variables to our ice cream sales prediction model (e.g., humidity, wind speed). $R^2$ might go up, but some of these variables may not actually improve predictions much.
- **Adjusted $R^2** accounts for this and could even decrease if the extra variables are not useful, helping us avoid overfitting.

---

### **Summary**:
- **Use $R^2$** when we want a general sense of how well our model explains the variation in the data.
- **Use Adjusted $R^2$** when we want a more reliable measure that doesn’t give extra credit for adding more variables unnecessarily. It’s especially useful when we have multiple predictors in our model.

In simple terms, **Adjusted $R^2$** is a "stricter" version of **$R^2$**, making sure we don't add too many variables just to make our model look better!

### **Elastic Net Regularization**

Elastic Net is a technique that combines the strengths of both **Ridge** and **Lasso** regression. It was developed to address a limitation of Lasso, where the selection of variables can be highly sensitive to the data, making it less stable. Elastic Net provides a balance by using both **L2** (from Ridge) and **L1** (from Lasso) penalties, making the model more robust in situations where many features are correlated.

The new objective function for Elastic Net looks like this:

$$RSS_{new} = RSS + \lambda_1 \sum(coeff^2) + \lambda_2 \sum(|coeff|)$$

- **RSS**: Residual Sum of Squares (the error term).
- **$\lambda_1$**: Controls the amount of **L2** regularization (Ridge penalty), which shrinks coefficients and handles multicollinearity.
- **$\lambda_2$**: Controls the amount of **L1** regularization (Lasso penalty), which performs feature selection by forcing some coefficients to zero.

In summary, Elastic Net combines both regularization techniques to stabilize variable selection and handle multicollinearity, while still shrinking coefficients to avoid overfitting.

### **Generalized Additive Models (GAMs)**

Generalized Additive Models (GAMs) are a flexible extension of linear models that allow us to handle non-linear relationships while keeping the model interpretable. GAMs differ from traditional linear models, like Linear Regression, by allowing the relationships between the features and the outcome to be more complex, but still additive.

In a standard linear model, the outcome is expressed as a simple weighted sum of the features:
$$y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_pX_p$$

In a **GAM**, instead of using a linear weighted sum like $\beta_1X_1$, we use flexible functions of the features, such as $f_1(X_1), f_2(X_2)$, which can capture non-linear patterns:
$$y = \beta_0 + f_1(X_1) + f_2(X_2) + \dots + f_p(X_p)$$

### Key Points:
- **Non-linearity with Interpretability**: GAMs allow each feature to have its own non-linear relationship with the outcome, making it more powerful than a simple linear model, while still retaining interpretability.

- **Additivity**: Even though GAMs allow non-linear relationships, they still model the outcome as the sum of individual feature effects. This makes it easy to interpret how each feature affects the target while keeping other variables constant.

### Example:
Imagine we are predicting house prices based on variables like **square footage** and **age of the house**. While linear models would assume a straight-line relationship (e.g., doubling square footage doubles the price), a GAM can model the more realistic non-linear relationships—perhaps house prices increase quickly for small houses but level off for larger ones. At the same time, we can still interpret the effect of each feature independently.

To learn more, explore the [pyGAM documentation](https://pygam.readthedocs.io/en/latest/notebooks/tour_of_pygam.html).

### **Model Ensembling**

### **Voting Regressor**

A **Voting Regressor** is an ensemble learning technique that combines multiple regression models to improve overall predictive performance. It works by fitting several base regressors (e.g., Linear Regression, Decision Trees, or any other regression models) on the entire dataset. Once each regressor has made its prediction, the **Voting Regressor** averages these individual predictions to generate a final output.

This method helps to balance the strengths of different models, reducing the impact of weaknesses in any single model and improving the accuracy and stability of predictions.

In simple terms, it’s like asking multiple experts for their opinions and then taking the average to get a more reliable prediction.

### Example:
Let’s say we’re predicting house prices using three models: Linear Regression, Random Forest, and K-Nearest Neighbors (KNN). Each model will predict a price for the same house, and the Voting Regressor will take the average of those predictions, resulting in a more robust final prediction.



### **Model Stacking**

**Model Stacking**, also known as **Stacked Generalization**, is a powerful ensemble learning technique that combines the predictions of multiple machine learning models to improve overall performance. Instead of relying on a single model, stacking leverages the strengths of different models by using a second model, known as the **meta-model**, to learn how to best combine the predictions of the **base models**.

#### Key Concepts:
- **Base Models (Level-0 Models):** These are the first-layer models that are trained on the dataset. They can be distinct models, such as Linear Regression, Decision Trees, or Neural Networks, which make individual predictions on the same data.
- **Meta-Model (Level-1 Model):** This model learns to combine the predictions from the base models. It takes the outputs of the base models as input and is trained to make the final prediction.

#### How Stacking Works:
1. **Training Base Models:** Each base model is trained on the entire training dataset, and they each generate their own predictions.
2. **Generating Meta-Model Training Data:** The predictions from the base models are used as input features for the meta-model. Often, **k-fold cross-validation** is used to generate out-of-fold predictions from the base models, which ensures that the meta-model does not overfit to the base models.
3. **Training the Meta-Model:** The meta-model is trained on the out-of-fold predictions of the base models and learns how to best combine their outputs to make a more accurate final prediction.

#### Why Use Stacking?
- **Combining Different Strengths:** Stacking allows the use of different model types (e.g., a linear model, a tree-based model, and a neural network), each with its own strengths. This diversity helps reduce overfitting and improves predictive accuracy.
- **Better Generalization:** By combining multiple models, stacking can better generalize to unseen data, often outperforming any individual model.

#### Example Scenario:
Let’s say we are predicting house prices using three different models: Linear Regression, Random Forest, and XGBoost. Each model will predict the price of a house. The **meta-model** (e.g., another linear regression model or a more complex algorithm) will take the predictions from these three models and combine them to produce a final, more accurate prediction.

#### Stacking vs. Other Ensemble Methods:
- **Bagging:** Stacking uses different base models, whereas bagging (e.g., Random Forest) typically uses multiple instances of the same model trained on different data samples.
- **Boosting:** In boosting (e.g., Gradient Boosting), models are trained sequentially to correct the errors of the previous model, while stacking trains base models independently and combines their predictions via a meta-model.

#### Example Formula (for Regression):
If we have two base models, $f_1(x)$ and $f_2(x)$, and a meta-model $h(x)$, then the final prediction $\hat{y}$ would be:
$$ \hat{y} = h(f_1(x), f_2(x)) $$

In practice, we can have many base models, and the meta-model will learn how to weight or combine their predictions optimally.



### **Model Blending**

**Model Blending** is an ensemble machine learning technique that combines predictions from multiple models to improve overall performance. Similar to **Stacking Generalization**, blending aims to leverage the strengths of different models; however, it differs in how the meta-training data is constructed.

#### Key Features of Blending:
- **Single Holdout Set:** Unlike stacking, which typically employs k-fold cross-validation to generate meta-training data, blending utilizes a single holdout set, which is a small subset of the training data reserved for validation. This subset is used to generate predictions that are combined to create the input for the meta-model.
- **Test Data Integration:** Predictions from the test dataset are also generated and used to create the input for the meta-model. This allows blending to produce a more comprehensive understanding of how well the ensemble performs on unseen data.

#### How Blending Works:
1. **Splitting the Dataset:** The dataset is divided into a training set and a holdout set. The holdout set will not be used to train the base models but will be used to evaluate their performance.
2. **Training Base Models:** Multiple base models are trained on the training set. Each model generates predictions based on the training data.
3. **Creating Meta-Model Input:** The predictions made by the base models on the holdout set are collected to form the input features for the meta-model.
4. **Training the Meta-Model:** The meta-model is then trained using the predictions from the base models on the holdout set. This model learns how to optimally combine the predictions of the base models to improve accuracy.
5. **Generating Test Predictions:** The trained base models generate predictions for the test set, which are also used as input features for the meta-model to make final predictions.

#### Advantages of Blending:
- **Simpler Implementation:** Blending is often simpler to implement than stacking, as it does not require the complexity of cross-validation for the meta-training data.
- **Potentially Faster Training:** By using a single holdout set, blending can reduce the training time compared to stacking, which involves multiple rounds of cross-validation.
- **Effective Use of Training Data:** Since blending relies on a holdout set rather than multiple cross-validation folds, it can utilize more training data for the base models.

#### Example Scenario:
Consider a situation where we want to predict customer churn using three different models: Logistic Regression, Random Forest, and Support Vector Machine (SVM). In blending:
- We would reserve a portion of the training data as a holdout set.
- Each model would be trained on the remaining training data.
- Predictions would be made on the holdout set, and these predictions would form the input for a meta-model, which could be another logistic regression model.
- Finally, predictions for the test data would also be generated using the trained base models, and these predictions would be combined by the meta-model to produce the final output.

#### Summary:
In essence, model blending combines the strengths of various models by employing a straightforward approach to integrate their predictions using a holdout set, making it an effective and efficient ensemble method for improving predictive performance.
